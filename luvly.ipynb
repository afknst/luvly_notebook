```python
import numpy as np
from luvly import Luvly, benchmark
import matplotlib.pyplot as plt
```

```python
def coral(_res):
    for k in [200, 230, 260, 300]:
        print(f"{k}:\t{np.count_nonzero(_res[:, 0] >= k) / 10000}")
    print(f"mean:\t{np.mean(_res[:, 0])}\nstd:\t{np.std(_res[:, 0])}")
    print(f"min:\t{np.min(_res[:, 0])}\nmax:\t{np.max(_res[:, 0])}")
    fig, ax = plt.subplots()
    ax.hist(_res[:, 0], bins=20, density=True)
    plt.title('Coral')
    plt.show()


def pearl(_res):
    for k in [250, 300, 400, 500, 600]:
        print(f"{k}:\t{np.count_nonzero(_res[:, 1] >= k) / 10000}")
    print(f"mean:\t{np.mean(_res[:, 1])}\nstd:\t{np.std(_res[:, 1])}")
    print(f"min:\t{np.min(_res[:, 1])}\nmax:\t{np.max(_res[:, 1])}")
    fig, ax = plt.subplots()
    ax.hist(_res[:, 1], bins=20, density=True)
    plt.title('Pearl')
    plt.show()
```

```python
res = np.array([Luvly(78).keep_rolling(benchmark) for _ in range(10000)])
coral(res)
pearl(res)
```

```python
res1 = np.array([Luvly(108).keep_rolling(benchmark) for _ in range(10000)])
coral(res1)
pearl(res1)
```
